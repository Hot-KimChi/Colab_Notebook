```python
import os

input_dir = 'images'
target_dir = 'annotations/trimaps/'

input_img_paths = sorted(
    [
        os.path.join(input_dir, fname) for fname in os.listdir(input_dir)
        if fname.endswith('.jpg')
    ])

target_paths = sorted(
    [
        os.path.join(target_dir, fname) for fname in os.listdir(target_dir)
        if fname.endswith('.png') and not fname.startswith('.')
    ])
```

>### 입력과 타깃을 2개의 넘파이 배열로 로드 & 구분

```python
import numpy as np
import random

img_size = (200, 200)
num_imgs = len(input_img_paths)

print(num_imgs)
```

```python
## 입력경로와 타깃 경로가 동일한 순서를 유지하도록 두 명령에 같은 시드(1337)를 사용
random.Random(1337).shuffle(input_img_paths)
random.Random(1337).shuffle(target_paths)
```

>### 함수 define
- path to input image
- path to target

```python
 def path_to_input_image(path):
    return img_to_array(load_img(path, target_size=img_size))

## 레이블이 0, 1, 2가 되도록 1을 뺀다.
## 전체 이미지를 input_image에 float32 배열로 로드하고 타깃 마스크는 targets에 uint8로 로드
def path_to_target(path):
    img = img_to_array(load_img(path, target_size=img_size, color_mode='grayscale'))
    img = img.astype('uint8') - 1
    return img
```

>#### 전체 이미지를 아래와 같은 형태로 load
- input_imgs에 float32 배열로 load(3개 채널:RGB)
- 타깃 마스크는 target에 uint8로 load(1개 채널: 정수 레이블)



```python
input_imgs = np.zeros((num_imgs, ) + img_size + (3, ), dtype='float32')
print(input_imgs.shape)

targets = np.zeros((num_imgs, ) + img_size + (1, ), dtype='uint8')
print(targets.shape)
```

```python
## image / target to array
## 테이터를 훈련세트 구분
for i in range(num_imgs):
    input_imgs[i] = path_to_input_image(input_img_paths[i])
    targets[i] = path_to_target(target_paths[i])

num_val_samples = 1000
train_input_imgs = input_imgs[:-num_val_samples]
print(train_input_imgs.shape)

train_targets = targets[:-num_val_samples]
print(train_targets.shape)
```

```python
val_input_imgs = input_imgs[-num_val_samples:]
val_targets = targets[-num_val_samples:]
print(val_input_imgs.shape, val_targets.shape)
```

>### Model Define

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def get_model(img_size, num_classes):
    inputs = keras.Input(shape=img_size + (3, ))

    ## normalization
    x = layers.Rescaling(1./255)(inputs)

    ## 특성맵 크기에 영향을 미치지 않도록 padding ='same'
    ## 3번의 downsampling(Maxpooling --> stride: 공간 위치 정보 유지)
    x = layers.Conv2D(64, 3, strides=2, activation='relu', padding='same')(x)
    x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)

    x = layers.Conv2D(128, 3, strides=2, activation='relu', padding='same')(x)
    x = layers.Conv2D(128, 3, activation='relu', padding='same')(x)

    x = layers.Conv2D(256, 3, strides=2, activation='relu', padding='same')(x)
    x = layers.Conv2D(256, 3, activation='relu', padding='same')(x)

    ## 3번의 upsampling(Conv2DTranspose)
    x = layers.Conv2DTranspose(256, 3, activation='relu', padding='same')(x)
    x = layers.Conv2DTranspose(256, 3, strides=2, activation='relu', padding='same')(x)

    x = layers.Conv2DTranspose(128, 3, activation='relu', padding='same')(x)
    x = layers.Conv2DTranspose(128, 3, strides=2, activation='relu', padding='same')(x)

    x = layers.Conv2DTranspose(64, 3, activation='relu', padding='same')(x)
    x = layers.Conv2DTranspose(64, 3, strides=2, activation='relu', padding='same')(x)

    outputs = layers.Conv2D(num_classes, 3, activation='softmax', padding='same')(x)

    model = keras.Model(inputs, outputs)

    return model
```

>### model 형태 확인

```python
model = get_model(img_size=img_size, num_classes=3)
model.summary()
```

```python
keras.utils.plot_model(model, show_shapes=True)
```

>### Model 분석
- 처음 절반: 이미지 분류, 컨브넷
    - Conv2D층을 쌓고 점진적 필터 개수 증가
    - 이미지 절반 downsampling(stride).
        - MaxPooling(분류작업): 풀링 윈도우 안 위치 정보 삭제 / 최대값만 반환
        - stride: 위치 정보 유지
    - 이미지를 작은 특성 맵으로 인코딩

- 나머지 절반: Conv2DTranspose층(upsampling)
    - 최종 출력은 타깃 마스크의 크기와 동일(200, 200, 3)

>### 모델 compile and 훈련

```python
epochs = 50

callbacks = [
    keras.callbacks.ModelCheckpoint('oxford_segmentation.keras', save_best_only=True)
            ]

model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_input_imgs, train_targets,
                    epochs=epochs,
                    callbacks=callbacks,
                    batch_size=64,
                    validation_data=(val_input_imgs, val_targets))
```

>### 훈련과 검증 손실 그래프 확인

```python
epochs = range(1, len(history.history['loss']) + 1)

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs, loss, 'bo', label='Training_loss')
plt.plot(epochs, val_loss, 'b', label='validation_loss')
plt.title('Training and validation loss')
plt.legend()
plt.figure()


accuracy = history.history["accuracy"]
val_accuracy = history.history["val_accuracy"]

plt.plot(epochs, accuracy, "bo", label="Training accuracy")
plt.plot(epochs, val_accuracy, "b", label="Validation accuracy")
plt.title("Training and validation accuracy")
plt.legend()
plt.show()
```
